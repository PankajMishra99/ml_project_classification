```python
df_copy=df.copy()
df_copy
```

```python
numeric_column = [col for col in df_copy.columns if df[col].dtype != 'O']
cat_col = [col for col in df_copy.columns if df[col].dtype=='O']
print("Numerical col", numeric_column)
print('=' * 50 )
print('Catgeorical column',cat_col)
```

```python
for col in numeric_column:
    sns.histplot(data=df_copy,x=col, kde=True,hue='type')
    plt.title(f"{col} vs type ")
    plt.xlabel('type')
    plt.ylabel(col)
    plt.show()
```
